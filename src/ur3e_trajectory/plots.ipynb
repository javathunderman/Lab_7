```python
import matplotlib.pyplot as plt
import pandas as pd

square_table = pd.read_csv("square.csv")
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(xs=square_table['x'], ys=square_table['y'], zs=square_table['z'])
plt.show()
```

```python
circle_table = pd.read_csv("circle.csv")
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(xs=circle_table['x'], ys=circle_table['y'], zs=circle_table['z'])
plt.show()
```
